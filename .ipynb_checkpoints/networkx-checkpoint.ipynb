```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G=nx.Graph()
```

```python
G.add_edge(1,2)
```

```python
G.add_edge(2,3,weight=0.9)
```

```python
G.add_edge("A","B")
G.add_edge("B","B")
G.add_node("C")
G.add_node(print)
```

```python
nx.draw_spring(G,with_labels=True)
plt.show()
```

```python
A=[(1,2),("A",2),(3,4),(4,6),(6,1)]
b=nx.from_edgelist(A)
```

```python
nx.draw_spring(b,with_labels=True)
plt.show()
```

```python
nx.draw_circular(G,with_labels=True)
```

```python
nx.draw_shell(G,with_labels=True)
```

```python
import numpy as np
```

```python

aa=nx.DiGraph()
aa.add_edge(1,2,weight=2)
nx.draw_spring(aa,with_labels=True)
plt.show()
```

```python
edge_list=[(1,2),(2,3),(3,4),(3,5),(4,6),(6,7),(2,8),(8,9),(9,4)]
GG=nx.Graph()
GG.add_edges_from(edge_list)
```

```python
nx.draw_planar(GG,with_labels=True)
```

```python
print(nx.shortest_path(GG,2,4))
print(GG.degree)
print(nx.degree_centrality(GG))
print(nx.closeness_centrality(GG))
```
